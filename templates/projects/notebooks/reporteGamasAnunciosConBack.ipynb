```python
for cat in cats:

    data = dev.maxiPublicaGamasTest.find_one({'categoryId':cat})
    backOffice = prod.pricesProcessed.find_one({'category_id':cat})


    # backOffice['avg_price_sale']['total']['y'][24] = 110000
    # backOffice['avg_price_sale']['total']['y'][29] = 130000

    frames = []
    suave = pd.Series(data = data['y'], index = data['x'], name='suave')
    suave.to_frame()


    backOffice = pd.Series(data = backOffice['avg_price_sale']['total']['y'],
                           index = backOffice['avg_price_sale']['total']['x'],
                           name='backOffice')
    backOffice = backOffice.to_frame()



    gam = pd.Series(data = data['sources']['gam']['y'], index = data['sources']['gam']['x'], name='gam')
    gam = gam.to_frame()
    ebc = pd.Series(data=data['sources']['ebc']['y'], index=data['sources']['ebc']['x'], name='ebc')
    ebc = ebc.to_frame()
    gdp_u = pd.Series(data=data['sources']['gdp_unique_ads']['y'], index = data['sources']['gdp_unique_ads']['x'], name='backOffice')
    gdp_u = gdp_u.to_frame()
    gdp_u_2 = pd.Series(data=data['sources']['gdp_uniqueAds_2']['y'], index=data['sources']['gdp_uniqueAds_2']['x'], name='backOffice')
    gdp_u_2 = gdp_u_2.to_frame()

    frames.append(suave)
    frames.append( gdp_u_2)
    frames.append(gdp_u)
    frames.append(ebc)
    frames.append(gam)
    frames.append(backOffice)



    frameAvg = pd.concat(frames, axis=0)
    frame = pd.concat(frames, axis=1)
    frame.index = pd.to_datetime(frame.index)
    frameAvg.index = pd.to_datetime(frameAvg.index)



    frame['dateSource'] = frame.index
    frameAvg['dateSource'] = frameAvg.index
    frame['categoryId'] = cat
    categ = frame.pop('categoryId')
    frame.insert(0, 'categoryId', categ)
    frame['PromedioAnuncios/B'] = frameAvg.groupby('dateSource').agg({'backOffice': pd.Series.mean})
    frame = frame.sort_values('dateSource')
    # frame = frame.astype({'PromedioAnuncios/B':'int'})

    range = pd.date_range(start= frame.index[0], end= frame.index[-1], freq='MS')
    frame = frame.reindex(range, fill_value='nan')

    frame['dateSource'] = frame.index
    frame['categoryId'] = cat

    frame['brand'] = data['brand']
    frame['model'] = data['model']
    brand = frame.pop('brand')
    model = frame.pop('model')
    frame.insert(1, 'brand', brand)
    frame.insert(2, 'model', model)

    frame.drop(['suave', 'dateSource'], axis=1, inplace=True)


    frame['PromedioAnuncios/B']  = frame['PromedioAnuncios/B'].rolling(5, min_periods=1, center=True, closed='both').mean()
    frame['PromedioAnuncios/B']  = frame['PromedioAnuncios/B'].rolling(5, min_periods=1, center=True, closed='both').mean()

    frame.columns = ['categoryId', 'brand', 'model', 'sitios', 'anunciosMaxi', 'ebc', 'gam', 'backOffice', 'promedio']
    print(frame)
    frame['promedio'].fillna(value=0, inplace=True)
    # frame = frame.astype({'PromedioAnuncios/B':'int'})
    frame['promedio'].replace(0, np.nan, inplace=True)
    frame.replace(' ', 'nan', inplace=True)


    # frame.to_csv(f'/Users/Artema/Desktop/{cat}.csv', date_format='%Y-%m-%d')
    fig, ax = plt.subplots(figsize=(18,8), dpi=120)

    plt.title(cat)
    ax.plot(frame.index, frame['promedio'], color='orange', label='promedio' )
    ax.plot(frame.index, frame['sitios'], color='red', label='sitios' )
    ax.plot(frame.index, frame['anunciosMaxi'], color='black', label='anunciosMaxi' )
    ax.plot(frame.index, frame['backOffice'], color='purple', label='backOffice')
    ax.plot(frame.index, frame['ebc'], color ='blue', label='ebc' )
    ax.plot(frame.index, frame['gam'], color='gray', label='gam' )
    plt.legend()

    final = final.append(frame)

    plt.savefig(f'/Users/Artema/Desktop/ReporteGamas/versiones 2019 back - noBack/imaÃÅgenes/conBack/{cat}.jpg')
    # fig.autofmt_xdate()
    plt.show()

```
